```python
import pandas as pd

melb_data_fe = pd.read_csv('D:\IDE\PY_10_Введение в Pandas\Data\melb_data_fe.csv')
melb_data_fe.head()
```

```python
melb_data_fe.info()
```

```python
melb_data_fe['Date'] = pd.to_datetime(melb_data_fe['Date'], dayfirst=True)
melb_data_fe['Date']
```

```python
quarter_sold = melb_data_fe['Date'].dt.quarter
quarter_sold.value_counts().iloc[1]
```

```python
exclude_list = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']
max_unique_count = 150
for col in melb_data_fe:
    if (melb_data_fe[col].nunique() < max_unique_count) and (col not in exclude_list):
        melb_data_fe[col] = melb_data_fe[col].astype('category')
display(melb_data_fe.info())
```

```python
melb_data_fe.sort_values(by=['Price', 'Distance']).loc[::10, ['Distance', 'Price']]
```

```python
mask1 = melb_data_fe['AreaRatio'] < -0.8
mask2 = melb_data_fe['Type'] == 'townhouse'
mask3 = melb_data_fe['SellerG'] == 'McGrath'
melb_data_fe[mask1 & mask2 & mask3].sort_values(
    by=['Date', 'AreaRatio'],
    ascending=[True, False],
    ignore_index=True
).loc[:, ['Date', 'AreaRatio']]
```

```python
melb_data_fe.sort_values(by='AreaRatio', ascending=False, ignore_index=True).loc[1558, 'BuildingArea']
```

```python
mask4 = melb_data_fe['Rooms'] > 2
melb_data_fe[mask2 & mask4].sort_values(
    by=['Rooms', 'MeanRoomsSquare'],
    ascending=[True, False],
    ignore_index=True
).loc[18, 'Price']
```
